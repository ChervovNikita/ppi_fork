```python
import numpy as np
def generate(data, t):
    for _ in range(t):
        ind1 = np.random.randint(len(data))
        ind2 = np.random.randint(len(data))
        yield (data[ind1], data[ind2], np.random.uniform(0, 1))
```

```python
result = []
result_test = []
for a, b, c in generate(real_data, 1000):
    result.append((a, b, c))

for a, b, c in generate(real_data, 200):
    result_test.append((a, b, c))
```

```python
import numpy as np

def generate_file(pairs, path):
    data = [['', '', row[0], '', '', row[1], row[2]] for row in pairs]
    np.random.shuffle(data)
    data = np.array(data, dtype='<U12')
    np.save(path, data)
```
